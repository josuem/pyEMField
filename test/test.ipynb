```python
import matplotlib.pyplot as plt
import numpy as np
import scienceplots

# plt.style.use(["science", "notebook"])

# %matplotlib qt

ax = plt.figure().add_subplot(projection='3d')

# Make the grid
x, y, z = np.meshgrid(np.arange(-1, 1.5, 0.5),
                      np.arange(-1, 1.5, 0.5),
                      np.arange(-1, 1.5, 0.5))

# Make the direction data for the arrows

def vectorial_field(x, y, z):
	v = 1 / (x**2 + y**2 + z**2)**(3/2) * [x, y, z]
	return v
	

field = vectorial_field(x, y, z)
u=field[0]
v=field[1]
w=field[2]

ax.quiver(x, y, z, u, v, w, length=0.2, normalize=False, 
		  color="black", linewidth=0.8)

# Agregar una esfera en el centro

u_sphere, v_sphere = np.mgrid[0:2*np.pi:20j, 0:np.pi:10j]

radio = 0.05
x_sphere = radio * np.cos(u_sphere) * np.sin(v_sphere)
y_sphere = radio * np.sin(u_sphere) * np.sin(v_sphere)
z_sphere = radio * np.cos(v_sphere)

ax.plot_surface(x_sphere, y_sphere, z_sphere, color='red')

elevation = 30  # Ángulo de elevación
azimuth = 30    # Ángulo de azimut
ax.view_init(elevation, azimuth)

ax.set(xlabel="x", ylabel="y", zlabel="z")
ax.set_title('Campo vectorial +Q')
plt.tight_layout()
plt.show()	
```
